```python
import numpy as np   #to do scientific calculation
import pandas as pd  #to access csv file 
import matplotlib.pyplot as plt #to represent the data
```

```python
#reading csv file
dataset = pd.read_csv("breast_cancer_dataset.csv")
dataset.head()     #showing top 5 rows
```

```python
#spliting the dataset into independent variables and dependent variable
X = dataset.iloc[:,2:32].values
Y = dataset.iloc[:,1].values
```

```python
#we need numerical data only so we need to change alphabets(M,B) into zeros and ones
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
Y = labelencoder.fit_transform(Y)
```

```python
#spliting the data into test sett and training set 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15,random_state=0)
```

```python
#feature scaling 
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.fit_transform(X_test)
```

```python
def logistic_regression(w, b, x_train, y_train, learning_rate,number_of_iterations):
    costs = []
    m = x_train.shape[1]    #it is for scaling
    #finding the cost and gradients and improving it with backpropogation
    for i in range(number_of_iterations):
        #forward propogration to find cost and gradients
        z = np.dot(w.T,x_train) + b   #finding z ehich is the prediction of our model
        yhat = 1/(1+np.exp(-z))    #taking the sigmoid function so the output will come in the range of 0<=z<=1
        loss = -y_train*np.log(yhat)-(1-y_train)*np.log(1-yhat) #calculating the loss to find the cost
        cost = (np.sum(loss))/m      #cost function we need to minimize it
        # backward propagation to coreect the parameters
        dw = (np.dot(x_train,((yhat-y_train).T)))/m   #it is the slope of w with respect to cost function
        db = np.sum(yhat-y_train)/m   #it is the slope of b with respect to cost function
        
        costs.append(cost)
        #updating the parameters 
        w = w - learning_rate*dw
        b = b - learning_rate*db
        if i % 2 == 0: 
            print ("Cost after iteration %i: %f" %(i, cost))
    
    # updated paramters 'w' and 'b'
    parameters = {"weight": w,"bias": b}
    
    return parameters, costs
```
